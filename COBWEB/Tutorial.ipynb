# Gain estimation

$\mathbb{E}_{t}[p_{t+1}] = \rho \mu + (1-\rho) p_{t}$, with $\mu$ the under. 

### Gain estimation with the coupling method

```python
gains_REE_coupling = {}
stds_REE_coupling = {}
for rho in tqdm(np.linspace(0,1,num=11)):
    agent = PoissonAgent(method="Weighted",mu=5/3, rho=rho)
    gains_REE_coupling[rho], stds_REE_coupling[rho], cost = agent.gain_tol_interval(max_restarts=200, num_runs=30, L=10,k=10,l=50, plot=True)
    print(f"Average number of timesteps per run = {cost/30}")
    print(f"Estimated gain for rho={round(rho,2)} is {gains_REE_coupling[rho]}")
```

```python
gains_over_coupling = {}
stds_over_coupling = {}
for rho in tqdm(np.linspace(0,1,num=11)):
    agent = PoissonAgent(method="Weighted",mu=8/3, rho=rho)
    gains_over_coupling[rho], stds_over_coupling[rho], cost = agent.gain_tol_interval(max_restarts=200, num_runs=30, L=10,k=10,l=50, plot=False)
    print(f"Average number of timesteps per run = {cost/30}")
    print(f"Estimated gain for rho={round(rho,2)} is {gains_over_coupling[rho]}")
```

```python
gains_under_coupling = {}
stds_under_coupling = {}
for rho in tqdm(np.linspace(0,1,num=11)):
    agent = PoissonAgent(method="Weighted",mu=2/3, rho=rho)
    gains_under_coupling[rho], stds_under_coupling[rho], cost = agent.gain_tol_interval(max_restarts=200, num_runs=30, L=10,k=10,l=50, plot=False)
    print(f"Average number of timesteps per run = {cost/30}")
    print(f"Estimated gain for rho={round(rho,2)} is {gains_under_coupling[rho]}")
```

```python
fig, ax = plt.subplots(figsize = (7,7))

# Extract x and y values from the 'gains' dictionary
x_values = list(gains_REE_coupling.keys())
y_values_overestimated = list(gains_over_coupling.values())
y_values_underestimated = list(gains_under_coupling.values())
y_values_REE = list(gains_REE_coupling.values())

# Extract standard deviation values from the 'std_deviations' dictionary
std_values_overestimated = list(stds_over_coupling.values())
std_values_underestimated = list(stds_under_coupling.values())
std_values_REE = list(stds_REE_coupling.values())

# Convert the dictionary of standard deviations to arrays
std_upper_REE = np.array(y_values_REE) + 2 * np.array(std_values_REE)
std_lower_REE = np.array(y_values_REE) - 2 * np.array(std_values_REE)

# Convert the dictionary of standard deviations to arrays
std_upper_overestimated = np.array(y_values_overestimated) + 2 * np.array(std_values_overestimated)
std_lower_overestimated = np.array(y_values_overestimated) - 2 * np.array(std_values_overestimated)

# Convert the dictionary of standard deviations to arrays
std_upper_underestimated = np.array(y_values_underestimated) + 2 * np.array(std_values_underestimated)
std_lower_underestimated = np.array(y_values_underestimated) - 2 * np.array(std_values_underestimated)

# Plotting
ax.plot(x_values, y_values_REE, label = 'REE', color='navy')
ax.fill_between(x_values, std_lower_REE, std_upper_REE, alpha=0.2, color = 'navy')

ax.plot(x_values, y_values_overestimated, label= 'Overestimation equilibrium price', color='red')
ax.fill_between(x_values, std_lower_overestimated, std_upper_overestimated, alpha=0.2, color='red')

ax.plot(x_values, y_values_underestimated, label='Underestimation equilibrium price', color='green')
ax.fill_between(x_values, std_lower_underestimated, std_upper_underestimated, alpha=0.2, color='green')
ax.legend()
ax.set_xlabel('Weight of expected equilibrium price (rho)')
ax.set_ylabel('Gain Estimate')
ax.set_title('Gain as function of weight of expected equilibrium price (Coupling Method)')
```

### Running one very long chain to see average reward after 1000 steps

If the slope of the supply function is bigger than that of the demand, the process diverges. It becomes periodic though, something our estimator can not deal with. 

Rational expectations

```python
cobweb = Cobweb(p_competitive=0,p_monopoly=5,demand_intercept=10,demand_slope=2,supply_intercept=5, supply_slope=1)
s_t = np.random.choice(cobweb.S)
cobweb.reset(s_t)

num_steps = 10_000
phi_0 = 5/3
phi_1 = 0

def policy(s_t):
    shock = np.random.binomial(2, 0.2) * np.random.choice([1,-1]) * (cobweb.S[1]-cobweb.S[0])
    a_t = phi_0 + phi_1 * s_t + shock
    a_t = min(cobweb.S, key=lambda num: abs(num - a_t))
    return a_t

means = []
for run in range(30):
    rewards = []
    expectations = []
    prices = []
    for t in range(1,num_steps):
        # Deciding on the action
            a_t = policy(s_t)
            expectations.append(a_t)
            # Update the timestep after taking the action
            # Take a step in the environment
            s_t, r = cobweb.step(a_t)
            prices.append(s_t)
            rewards.append(r)
    means.append(np.mean(rewards[5_000:]))

print(f"Average reward = {np.mean(means)}")
```

Naive expectations

```python
cobweb = Cobweb(p_competitive=0,p_monopoly=5,demand_intercept=10,demand_slope=2,supply_intercept=5, supply_slope=1)
s_t = np.random.choice(cobweb.S)
cobweb.reset(s_t)

num_steps = 10_000
phi_0 = 0
phi_1 = 1

def policy(s_t):
    shock = np.random.binomial(2, 0.2) * np.random.choice([1,-1]) * (cobweb.S[1]-cobweb.S[0])
    a_t = phi_0 + phi_1 * s_t + shock
    a_t = min(cobweb.S, key=lambda num: abs(num - a_t))
    return a_t

means = []
for run in range(30):
    rewards = []
    expectations = []
    prices = []
    for t in range(1,num_steps):
        # Deciding on the action
            a_t = policy(s_t)
            expectations.append(a_t)
            # Update the timestep after taking the action
            # Take a step in the environment
            s_t, r = cobweb.step(a_t)
            prices.append(s_t)
            rewards.append(r)
    means.append(np.mean(rewards[5_000:]))

print(f"Average reward = {np.mean(means)}")
```
